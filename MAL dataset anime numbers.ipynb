```python
import pandas as pd
```

```python
path = "C:/Users/moham/Downloads/Anime dataset"
```

```python
all_anime = pd.read_csv(r"C:\Users\moham\Downloads\Anime dataset\anime-dataset-2023.csv")
```

```python
all_anime.drop(columns="Synopsis", inplace=True)
```

```python
all_anime[["Season", "Year"]] = all_anime["Premiered"].str.split(' ', expand = True)
```

```python
all_anime_tv = all_anime[(all_anime["Type"] == "TV")]
```

```python
all_anime.drop(columns="Premiered", inplace=True)
```

```python
all_anime["Score"] = all_anime["Score"].str.replace("UNKNOWN", "None")
```

```python
all_anime.drop(all_anime[all_anime['Score'] == 'None'].index, inplace=True)
```

```python
all_anime.drop(all_anime[all_anime['Season'] == 'UNKNOWN'].index, inplace=True)
```

```python
all_anime.drop(all_anime[all_anime['Year'] == 'None'].index, inplace=True)
```

```python
all_anime.drop(all_anime[all_anime['Rank'] == 'UNKNOWN'].index, inplace=True)
```

```python
all_anime.dtypes
```

```python
all_anime["Scored By"] = all_anime["Scored By"].apply(lambda x: float(x))
```

```python
all_anime.drop(all_anime[all_anime['Episodes'] == 'UNKNOWN'].index, inplace=True)
```

```python
anime_per_season = all_anime.groupby("Season").agg({"anime_id": "count" ,"Score": "mean", "Rank": "mean"}).sort_values(by="Score", ascending=False)
```

```python
all_anime["Rank"] = all_anime["Rank"].apply(lambda x: int(x))
```

```python
anime_per_year = all_anime.groupby("Year").agg({"anime_id": "count" ,"Score": "mean", "Rank": "mean"}).sort_values(by=["Score", "anime_id"], ascending=[False, True])
```

```python
anime_per_year.to_csv("animesPerYear.csv")
```

```python
anime_per_season.to_csv(path + "AnimesPerseason.csv")
```
